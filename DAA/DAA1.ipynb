```python
def fibonacci_recursive(n):
    global step_count
    step_count += 1
    
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


def fibonacci_iterative(n):
    step_count = 0
    
    if n == 0:
        print(0)
    elif n == 1:
        print(0)
        print(1)
    else:
        current = 0
        p1 = 1
        p2 = 2
        
        print(current)
        print(p1)
        print(p2)
        
        for _ in range(2, n - 1):
            current = p1 + p2
            print(current)
            step_count += 1
            p1 = p2
            p2 = current

    print("Total steps (iterations):", step_count)


def main():
    while True:
        print("\nChoose Method:")
        print("1. Recursive (get nth Fibonacci number)")
        print("2. Iterative (print Fibonacci series)")
        print("3. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == '1':
            n = int(input("Enter the position (n): "))
            global step_count
            step_count = 0
            result = fibonacci_recursive(n)
            print(f"Fibonacci({n}) = {result}")
            print("Total steps (recursive calls):", step_count)
        
        elif choice == '2':
            n = int(input("Enter number of terms: "))
            fibonacci_iterative(n)
        
        elif choice == '3':
            print("Exiting...")
            break
        
        else:
            print("Invalid choice, please try again.")


if __name__ == "__main__":
    step_count = 0
    main()
```
