```python
class Item:
    def __init__(self, profit, weight):
        self.profit = profit
        self.weight = weight
        self.ratio = profit / weight  # profit-to-weight ratio


def sort_by_ratio(items):
    # Sort in descending order by ratio
    items.sort(key=lambda x: x.ratio, reverse=True)


def fractional_knapsack(W, items):
    sort_by_ratio(items)

    total_profit = 0.0
    current_weight = 0

    for item in items:
        if current_weight + item.weight <= W:
            # Take whole item
            current_weight += item.weight
            total_profit += item.profit
        else:
            # Take fractional part
            remain = W - current_weight
            total_profit += item.ratio * remain
            break

    return total_profit


# ---- Main Program ----

n = int(input("Enter number of items: "))

items = []
for i in range(n):
    profit = int(input(f"Enter profit of item {i+1}: "))
    weight = int(input(f"Enter weight of item {i+1}: "))
    items.append(Item(profit, weight))

W = int(input("Enter capacity of knapsack: "))

max_profit = fractional_knapsack(W, items)
print(f"Maximum profit = {max_profit:.2f}")
```
