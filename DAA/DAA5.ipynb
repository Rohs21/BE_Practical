```python
import random

def quick_sort(arr, pivot_type='last'):
    if len(arr) <= 1:
        return arr

    if pivot_type == 'random':
        pivot_index = random.randint(0, len(arr) - 1)
        pivot = arr[pivot_index]
        left = [x for i, x in enumerate(arr) if x <= pivot and i != pivot_index]
        right = [x for i, x in enumerate(arr) if x > pivot and i != pivot_index]
    else:  # default: last element pivot
        pivot = arr[-1]
        left = [x for x in arr[:-1] if x <= pivot]
        right = [x for x in arr[:-1] if x > pivot]

    return quick_sort(left, pivot_type) + [pivot] + quick_sort(right, pivot_type)


def get_array_from_user():
    while True:
        user_input = input("Enter numbers separated by spaces: ").strip()
        if not user_input:
            print("Input cannot be empty. Try again.")
            continue
        try:
            return [int(x) for x in user_input.split()]
        except ValueError:
            print("Please enter valid integers only.")


def main():
    while True:
        print("\nChoose an option:")
        print("1 - Quick Sort with random pivot")
        print("2 - Quick Sort with last element pivot")
        print("3 - Exit")

        choice = input("Enter your choice (1/2/3): ").strip()

        if choice == '3':
            print("Exiting program. Goodbye!")
            break
        elif choice in ['1', '2']:
            arr = get_array_from_user()
            pivot_type = 'random' if choice == '1' else 'last'
            sorted_arr = quick_sort(arr, pivot_type)
            print("Sorted array:", sorted_arr)
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")


if __name__ == "__main__":
    main()
```
